```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------------------------
#  Create or Load the Dataset
# -----------------------------------------------------------

# Create a sample dataset
data = {
    'PlotArea': [2000, 3000, 2500, 4000, 3500, 4500, 2800, 3700, 2200, 3100],
    'PlotConfig': ['Corner', 'Inside', 'Corner', 'Cul-de-sac', 'Inside',
                   'Corner', 'Inside', 'Cul-de-sac', 'Corner', 'Inside'],
    'Condition': [8, 6, 7, 9, 5, 8, 6, 7, 8, 6],
    'YearBuilt': [2005, 1998, 2010, 2015, 2000, 2012, 2008, 2014, 2011, 2009],
    'YearRemodAdd': [2010, 2005, 2012, 2018, 2003, 2015, 2010, 2016, 2013, 2011],
    'Location': ['Hyderabad', 'Chennai', 'Delhi', 'Bangalore', 'Pune',
                 'Hyderabad', 'Mumbai', 'Chennai', 'Delhi', 'Pune'],
    'PricePerSqFt': [4500, 4000, 4800, 5200, 4100, 5500, 5000, 4600, 4700, 4300]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display dataset
print(" House Dataset:\n")
print(df)

# -----------------------------------------------------------
#  Statistical Summary
# -----------------------------------------------------------

print("\n Statistical Summary:")
print(df.describe())

# -----------------------------------------------------------
#  Data Exploration
# -----------------------------------------------------------

# Average price per location
avg_price_location = df.groupby('Location')['PricePerSqFt'].mean()
print("\n Average Price per Location (₹/sq.ft):\n", avg_price_location)

# -----------------------------------------------------------
#  Visualization using matplotlib
# -----------------------------------------------------------

# Plot 1: Condition vs Price per sq. ft
plt.figure(figsize=(8,5))
plt.bar(df['Condition'], df['PricePerSqFt'], color='skyblue')
plt.xlabel("Condition of House (1-10)")
plt.ylabel("Price per Sq.Ft (₹)")
plt.title("Condition vs Price per Sq.Ft")
plt.show()

# Plot 2: Plot Area vs Price per Sq. Ft
plt.figure(figsize=(8,5))
plt.scatter(df['PlotArea'], df['PricePerSqFt'], color='green')
plt.xlabel("Plot Area (sq.ft)")
plt.ylabel("Price per Sq.Ft (₹)")
plt.title("Plot Area vs Price per Sq.Ft")
plt.show()

# Plot 3: Average price by Location
plt.figure(figsize=(8,5))
plt.bar(avg_price_location.index, avg_price_location.values, color='orange')
plt.xlabel("Location")
plt.ylabel("Average Price per Sq.Ft (₹)")
plt.title("Average Price by City")
plt.xticks(rotation=45)
plt.show()

# -----------------------------------------------------------
#  Derived Metric: Total House Price
# -----------------------------------------------------------
df['TotalPrice'] = df['PlotArea'] * df['PricePerSqFt']

print("\n💰 Total Price of each House (in ₹):")
print(df[['Location', 'PlotArea', 'PricePerSqFt', 'TotalPrice']])
```
